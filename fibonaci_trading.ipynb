```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
import os
```

```python
diretorio=os.getcwd()
df=pd.read_csv(diretorio+"/"+"btc.csv",sep=";",index_col="datetime")
df.head(2)
```

```python
## Let`s identify the fibonacci ratio to a specific period
df1=df.close[-300:-220]
plt.plot(df1)
plt.show()
```

```python
p_max=df1.max()
p_min=df1.min()
diff=p_max-p_min
fibonacci1=p_max-0.236*diff   #less 23.6%
fibonacci2=p_max-0.382*diff   #less 38.2%
fibonacci3=p_max-0.618*diff   #less 61.8%
fibonacci4=p_max-0.786*diff   #less 78.6%
print('level','  ','Price')
print(0,"      ",p_max)
print(0.236,"  ",fibonacci1)
print(0.382,"  ",fibonacci2)
print(0.618,"  ",fibonacci3)
print(0.786,"  ",fibonacci4)
print(1,"      ",p_min)
```

```python
fig, ax = plt.subplots()
ax.plot(df1, color='black')
ax.axhspan(fibonacci1, p_min, alpha=0.4, color='lightsalmon')
ax.axhspan(fibonacci2, fibonacci1, alpha=0.5, color='palegoldenrod')
ax.axhspan(fibonacci3, fibonacci2, alpha=0.5, color='palegreen')
ax.axhspan(p_max, fibonacci3, alpha=0.5, color='powderblue')
plt.ylabel("Price")
plt.xlabel("Date")
plt.legend(loc=2)
plt.show()
```
